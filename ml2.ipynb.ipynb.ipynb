```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('student_marks.csv')
test_1 = df["Test_1"]
test_2 = df["Test_2"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)
plt.pie(test_2, labels=test_1, colors=colors)
plt.title("Students Marks")
plt.show()
```

```python
plt.scatter(test_1, test_2, s=100, alpha=0.6, edgecolor='black', linewidth=1)

plt.title('Students Marks')
plt.xlabel('Test_1')
plt.ylabel('Test_2')

plt.tight_layout()
plt.show()
```
